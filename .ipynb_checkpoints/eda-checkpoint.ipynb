```python
# Run this before executing: !pip install pandas pandas-profiling numpy matplotlib

# Here we use some "manual EDA and data visualizations"

# Import libraries
import io

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas_profiling as pp
```

```python
# Load datasets
ratings   = pd.read_csv('rating_final.csv', sep=';')
geoplaces = pd.read_csv('geoplaces2.csv', sep=';', encoding="ISO-8859-1") # Changed encoding to ISO-8859-1 to solve compile issue

# Some basic info about the datasets
ratings.info()
print('\n')
geoplaces.info()
```

```python
# Show the dfs
print(f'Ratings\n{ratings.head()}\n\nGeoplaces\n{geoplaces.head()}')
```

```python
geoplaces = geoplaces.replace('\?', 'unknown', regex=True) # Change all ? charaters to string unknown
```

```python
# Find duplicates of placeID on geoplaces, no need in ratings df due to that there can be multiple ratings per place by one user
print('Amount of duplicate placeIDs in geoplaces df:', len(geoplaces[geoplaces.duplicated(['placeID'], keep=False)]))

# Find any NULL values in both dfs
print(f'\nNULL Values\n{ratings.isnull().sum()} \n\n{geoplaces.isnull().sum()}')
```

```python
# Some descriptive numbers 

print('Ratings DF amount of rows, and cols', ratings.shape)
ratings.describe()
```

```python
print('Geoplaces DF amount of rows, and cols', geoplaces.shape)
geoplaces.describe()
```

```python
# We can also use pandas-profiling for some automatic and easy to use visualisations
pp.ProfileReport(ratings)
```

```python
pp.ProfileReport(geoplaces)
```

```python
# We do some more statistics and graphs
%matplotlib inline

matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (20, 10)
matplotlib.rcParams['figure.facecolor'] = '#00000000'

# Get the avg ratings
print(f'Avg rating: {int(ratings.rating.mean())}')
print(f'Avg food rating: {int(ratings.food_rating.mean())}')
print(f'Avg service rating: {int(ratings.service_rating.mean())}')
```

```python
# Print out amount of ratings per userID
print('Amount of ratings per userID')
ratings.userID.value_counts().plot(kind='bar')
```

```python
# Print out amount of ratings per placeID
print('Amount of ratings per placeID')
ratings.placeID.value_counts().plot(kind='bar')
print(f'Avg amount of placeID ratings: {int(ratings.placeID.value_counts().mean())}')
```
